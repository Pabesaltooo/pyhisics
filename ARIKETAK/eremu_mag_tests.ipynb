Karga Isolatua (q=36 nC) V — 4.5 X 107 m/s-ko abladuraz ari da mugitzen Y+ ardatzaren noranzkoan, x=3m lerroan zehar. Karga honek jatorrian sortutako eremu magnetikoa x=3m, y=4m eta z=0m puntuan dagoenean hauxe da:


---

Formula de un campo magnetico creado por una particula:

$$
\vec{B} = \frac{\mu_0}{2\pi} \frac{q\cdot \vec{v}\times\vec{r}}{r^2}
$$

```python
q = Quantity(3.6, 'nC')
v = Quantity([0,4.5e7, 0], 'm/s')
dist = Quantity([3,4,0],'m').magnitude()
r_vec = Quantity([-3,-4,0])

B = mu0 / (2*pi) * q * v.cross(r_vec) / dist**2
B
```

Elektroi bat pausagunetik azeleratu dugu eta, ondoren, B=0.00127 T-ko eremu magnetiko uniformea dagoen gune batera sartzen da, B eta •elektroiaren  abiadura elkarzutak direlarik. Elektroiak 5 ns irauten du B eremu magnetikoaren eraginpean. Kalkulatu elektroiaren sarrera-abiadura eta irteera-abiaduraren arteko angelua.

--- 

$$
\begin{aligned}
& \vec{F}_Z = \vec{F}_L\\

& m \cdot \vec{a_N} = q \, \vec{v} \times \vec{B} \\

\xRightarrow{\vec{v} \perp \vec{B}} \quad
& m \cdot a_N = q \, v B \\

& m \frac{{v_L}^2}{r} = qvB \\

& m \frac{v_L}{r} = qB \\

\xRightarrow{v = \omega r} \quad
& m \frac{\omega r}{r} = qB \\

& m \omega = qB \\

\underset{\frac{\theta}{t}=\omega \, \text{kte}}{\xRightarrow{\frac{d\omega}{dt} = \alpha = 0}} \quad
& m \frac{\theta}{t} = qB \\

& \boxed{\theta = \frac{q\,B\,t}{m}}

\end{aligned} 

$$

```python
B = Quantity(0.00127, 'T')
t = Quantity(5, 'ns')

tht = e_*B / (me_) * t

(tht)
```

Calculate the magnetic field and its direction at point P, which is 2.0 cm away
from the top wire and 4.0 cm from the bottom wire. Assume both wires are infinitely long and each carries a current of 0.15 A.



Biot-Savart (eroale zuzen infinitua):

$$
\vec{B} = \frac{\mu_0}{2\pi} \frac{I}{d} \hat u
$$

Gainezarmen printzipioa aplikatuz:

$$
\vec{B}_T = \sum \vec{B}_i = \vec{B}_1 + \vec{B}_2
$$

Kasu honetan norabide berdinekoak dira $P$ puntuan baina aurkako noranzkokoak intentsitatea dela eta.

$$
B = B_1 - B_2
$$

```python
I = Quantity(0.15, 'A')

d1 = Quantity(4, 'cm')
d2 = Quantity(2, 'cm')

B1 = mu0/(2*pi) * I / d1
B2 = mu0/(2*pi) * -I / d2

B_T = B1 + B2

(B_T)
```
